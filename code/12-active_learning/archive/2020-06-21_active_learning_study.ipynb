```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

## Histograms

```python
path = "/home/manuto/Documents/world_bank/bert_twitter_labor/code/twitter/data/jun3_10Klabels/data_binary_pos_neg_unbalanced/bert-base-cased-conversational_scores"
```

```python
for label in ["is_hired_1mo","is_unemployed","job_offer","job_search","lost_job_1mo"]:
    df = pd.read_csv(os.path.join(path,'val_{}.csv'.format(label)), lineterminator='\n')
    base_rate = (df.loc[df['labels']==1].shape[0]/df.shape[0])*100
    x_line = np.percentile(df['scores'],100 - base_rate)
    plt.hist(df['scores'], bins=30, log=True)
    plt.legend()
    #plt.hist(df.loc[df['labels']==0]['scores'], bins, log=True, color='red',label='negative')
    plt.axvline(x_line, color='red')
    plt.xlabel('Score value')
    plt.ylabel('Number of tweets (log scale)')
    print("************** Label {} **************".format(label))
    print('Base rate percentile: ', x_line)
    plt.tight_layout()
    plt.show()
```

```python
for label in ["is_hired_1mo","is_unemployed","job_offer","job_search","lost_job_1mo"]:
    df = pd.read_csv(os.path.join(path,'val_{}.csv'.format(label)), lineterminator='\n')
    base_rate = (df.loc[df['labels']==1].shape[0]/df.shape[0])*100
    x_line = np.percentile(df['scores'],100 - base_rate)
    plt.hist([df.loc[df['labels']==1]['scores'],df.loc[df['labels']==0]['scores']], bins=30, log=True, label=['positive','negative'])
    plt.legend()
    plt.axvline(x_line, color='red')
    plt.xlabel('Score value')
    plt.ylabel('Number of tweets (log scale)')
    print("************** Label {} **************".format(label))
    print('Base rate percentile: ', x_line)
    plt.tight_layout()
    plt.show()
```
