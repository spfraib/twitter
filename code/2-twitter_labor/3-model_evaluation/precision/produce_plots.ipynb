```python
from results import *
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
sns.set()
```

```python
def extract_data(label, data_dict):
    precision_list = list()
    for iter_nb in range(4):
        precision_list.append(data_dict[iter_nb][label])
    if 4 in data_dict.keys():
        precision_list.append(data_dict[4][label])
    return precision_list
```

```python
output_path = '/home/manuto/Documents/world_bank/bert_twitter_labor/twitter-labor-data/data/fig/precision'
```

```python
labels=['is_hired_1mo', 'is_unemployed', 'job_offer', 'job_search', 'lost_job_1mo']
```

```python
iter_nb = range(5)
```

# top 50

```python
for label in labels:
    fig,ax = plt.subplots(figsize=(4,4))
    precision_list_our_method = extract_data(label, our_method_top50_dict)
    precision_list_adaptive = extract_data(label, adaptive_top50_dict)
    ax.plot(iter_nb, precision_list_our_method, 'b-', label='our method')
    ax.plot(iter_nb, precision_list_adaptive, 'r-', label='adaptive retrieval')
    ax.tick_params(which='both',direction='in',pad=3)
#     ax.locator_params(axis='y',nbins=6)
    ax.set_ylabel('Precision (top50)',fontweight='bold')
    ax.set_xlabel('Iteration number',fontweight='bold')
    ax.set_title(label.replace('_',' ').replace('1mo','').title(),fontweight='bold')
    ax.legend(loc='best',fontsize=9)
    ax.set_ylim([-0.05,1.05])
    plt.savefig(os.path.join(output_path,f'precision_top50_{label}.png'),bbox_inches='tight', format='png' ,dpi=1200, transparent=False)
```

# top 10K

```python
for label in labels:
    fig,ax = plt.subplots(figsize=(4,4))
    precision_list_our_method = extract_data(label, our_method_top10K_dict)
    precision_list_adaptive = extract_data(label, adaptive_top10K_dict)
    precision_list_uncertainty = extract_data(label, uncertainty_top10K_dict)
    ax.plot(iter_nb, precision_list_our_method, 'b-', label='our method')
    ax.plot(iter_nb, precision_list_adaptive, 'r-', label='adaptive retrieval')
    ax.plot(range(4), precision_list_uncertainty, 'g-', label='uncertainty sampling')
    ax.tick_params(which='both',direction='in',pad=3)
#     ax.locator_params(axis='y',nbins=6)
    ax.set_ylabel('Precision (top10k)',fontweight='bold')
    ax.set_xlabel('Iteration number',fontweight='bold')
    ax.set_title(label.replace('_',' ').replace('1mo','').title(),fontweight='bold')
    ax.legend(loc='best',fontsize=9)
    ax.set_ylim([-0.05,1.05])
    plt.savefig(os.path.join(output_path,f'precision_top10K_{label}.png'),bbox_inches='tight', format='png' ,dpi=1200, transparent=False)
```
