```python
import pandas as pd
import os
import numpy as np
import math
```

```python
def build_has_description(x):
    if pd.isnull(x):
        return 0
    else:
        return 1
```

```python
for country_code in ['MX', 'BR']:
    path = f'/home/manuto/Downloads/data_{country_code}_users'
    predictions_df = pd.read_csv(os.path.join(path, f'Result_{country_code}_250_users'), sep='\t', header=None)
    predictions_df.columns = ['id', 'gender_predicted', 'age_predicted', 'org_predicted']
    user_info_df = pd.read_csv(os.path.join(path, f'User_Info_{country_code}.csv'))
    total_df = pd.merge(user_info_df, predictions_df, on=['id'])
    new_columns = ['gender_true', 'age_true', 'org_true', 'default_picture', 'has_description', 'real_name', 'real_picture', 'bot', 'last_active']
    for new_column in new_columns:
        if new_column == 'has_description':
            total_df[new_column] = total_df['description'].apply(build_has_description)
        else:
            total_df[new_column] = np.nan
    total_df = total_df.sample(n=100).reset_index(drop=True)
    total_df.to_csv(f"/home/manuto/Downloads/data_{country_code}_users/demographics_labelling_{country_code}.csv")
```
